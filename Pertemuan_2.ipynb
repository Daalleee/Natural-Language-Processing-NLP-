```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://megapolitan.kompas.com/read/2025/09/08/07550041/ada-tiga-titik-demo-di-jakarta-hari-ini-8-september-2025"
# Ambil halaman web
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
# Ambil semua teks dari paragraf <p>
paragraphs = soup.find_all("p")
text = " ".join([para.get_text() for para in paragraphs])
# Tampilkan hasil teks mentah
print("Teks Mentah (Scraping):\n")
wrapped_text = re.sub(r'\. ', '.\n', text)
print(wrapped_text)  # Menampilkan teks dengan baris baru setelah titik
```

```python
cleaned_text = preprocess_text(text)
print("Tokens setelah preprocessing:", cleaned_text[:50])  # Tampilkan 50 token pertama
```
