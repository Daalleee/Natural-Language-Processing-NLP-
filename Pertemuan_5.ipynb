```python
import nltk
from nltk import CFG, ChartParser
from nltk.parse import CoreNLPParser
import spacy
```

```python
nltk.download('punkt')
```

```python
sentences = { "id": [ "Ani memberi Budi sebuah buku .",
                     "Mahasiswa sedang belajar di kampus .",
                      "Jokowi pergi ke Bali dengan pesawat ." ],
               "en": [ "Ani gives Budi a book .",
                      "The student is studying at campus .",
                       "Jokowi goes to Bali by plane ." ] }
```

```python
grammar_id = CFG.fromstring("""
      S -> NP VP '.'
      NP -> PropN | Det N | N
      VP -> VP | V | V VP | AUX VP | V NP NP | V NP | V PP | V PP PP
      PP -> Preposisi NP
      PropN -> 'ani' | 'budi' | 'jokowi'
      Det -> 'sebuah' | 'seorang'
      N -> 'buku' | 'mahasiswa' | 'kampus' | 'pesawat' | 'bali'
      V -> 'memberi' | 'belajar' | 'pergi'
      Preposisi -> 'ke' | 'di' | 'dengan'
      AUX ->'sedang'
      PUNCT -> '.'
      """)
```

```python
parser = ChartParser(grammar_id)

for sent in sentences["id"]:
  tokens = sent.lower().split()
  print(f"\nKalimat: {sent}")
  try:
    for tree in parser.parse(tokens):
      print(tree)
      tree.pretty_print()
  except:
    print("Tidak dapat di-parse (aturan tidak lengkap)")
```

```python
nlp_en = spacy.load("en_core_web_sm")

for sent in sentences["en"]:
    doc = nlp_en(sent)
    print(f"\nKalimat: {sent}")

    for token in doc:
        print(f"{token.text} <-{token.dep_}-- {token.head.text}")

    # Visualisasi (opsional)
    from spacy import displacy
    displacy.render(doc, style="dep", jupyter=True)
```

```python
# Alternatif: Flair atau Stanza
import stanza
stanza.download('id')
nlp_id = stanza.Pipeline('id')
doc = nlp_id("Ani memberi Budi sebuah buku.")
for sentence in doc.sentences:
    for word in sentence.words:
        print(f"{word.text} -> {word.head} ({word.deprel})")
```
